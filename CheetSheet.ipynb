# 1. ML Foundations and Paradigms of Machine Learning

Here is a cheat sheet summarizing the essential formulas, relationships, rules, and concepts from the provided machine learning material:

### **1. Important Formulas & Metrics**

**Classification Metrics:**
These evaluate how well a model categorizes discrete data based on True Positives (TP), False Positives (FP), True Negatives (TN), and False Negatives (FN):
*   **Precision:** Measures exactness. 
    `Precision = TP / (TP + FP)`
*   **Recall:** Measures completeness. 
    `Recall = TP / (TP + FN)`
*   **F-score:** The weighted average of precision and recall. 
    `F-score = 2 * (Precision * Recall) / (Precision + Recall)`

**Regression Metrics:**
Used when predicting continuous values.
*   **Mean Squared Error (MSE):** `MSE = (1/n) * Σ(y - d)²` 
    *(Where `y` is the model's predicted output, `d` is the desired actual output, and `n` is the total number of instances)*.

**Supervised Learning Mapping:**
*   **Mathematical Function:** `y = f(x_i)` for `i = 1, ..., n` 
    *(Where `y` is the dependent variable/output, `x_i` are the independent variables/inputs, and `f` is the model's estimated function)*.

---

### **2. Important Rules & Relationships**

**Data Types Output Relationships:**
*   **Continuous Outputs** ➞ Use **Regression Models** (e.g., predicting house prices).
*   **Discrete/Categorical Outputs** ➞ Use **Classification Models** (e.g., spam vs. not spam).
*   **No Provided Outputs/Labels** ➞ Use **Clustering/Unsupervised Learning** (e.g., grouping customers by similarity).

**Model Fitting Rules:**
*   **The Complexity Trade-off:** There is a delicate balance between goodness of fit and model complexity. 
*   **Overfitting:** Happens when a model is too complex and has too many degrees of freedom. It learns the "noise" in the training data, leading to great training performance but poor generalization (failing on testing data).
*   **Underfitting:** Happens when the model lacks the complexity to capture the data's inherent structure, resulting in poor performance on *both* training and testing datasets.

**Data Partitioning:**
*   **Training Set:** Used exclusively to develop and build the mathematical model.
*   **Testing Set:** Used exclusively to measure and evaluate the performance/accuracy of the developed model.

---

### **3. Points to Remember: The 4 Machine Learning Paradigms**

**1. Supervised Learning**
*   **What it is:** Learns from a dataset containing **both inputs and desired outputs**.
*   **Goal:** To adjust parameters to minimize a loss function (like MSE for regression or misclassifications for classification) until the optimum setting is achieved.
*   **Common Techniques:** Linear regression, Support Vector Machines (SVM), Naïve Bayes, Decision trees, and K-nearest neighbors.

**2. Unsupervised Learning**
*   **What it is:** Learns from datasets containing **inputs only (no labels)**.
*   **Goal:** To discover hidden salient patterns, group similar instances (clustering), reduce dimensionality (removing irrelevant variables), and detect outliers/anomalies.
*   **Common Techniques:** K-Means, hierarchical clustering, DBSCAN, and Principal Component Analysis (PCA).

**3. Semi-Supervised Learning**
*   **What it is:** A hybrid approach used when you have a **few labeled instances and many unlabeled instances**.
*   **Goal:** Saves massive computational and manual labeling costs. It first classifies the labeled data, then uses unsupervised clustering to transfer those labels to similar unlabeled data points. 

**4. Reinforcement Learning**
*   **What it is:** A "cause and effect" approach for sequential decision-making where an **Agent** interacts with an **Environment**.
*   **Goal:** To find an action policy that maximizes a **Reward** over time through trial-and-error. 
*   **Key Distinction:** Unlike supervised learning, it relies on a feedback reward function rather than strict input/output mapping. It is widely used in robotic movement and game playing.

---

### **4. Foundational Concept**
**Traditional Programming vs. Machine Learning:** 
In traditional programming, you input data and a *program (rules)* into a computer to generate an output. In machine learning, you input data and the *desired outputs* into a computer, and the computer generates the *program (model)*.

# 2. ML Fundamentals of Clustering Analysis and K-Means Implementation

Here is a cheat sheet summarizing the essential formulas, relationships, rules, and concepts for clustering analysis based on the sources:

### **1. Core Concepts & Important Rules**
*   **Clustering Goal:** The objective is to group data records into natural groups based on similarity metrics. 
*   **The Golden Rule of Clustering:** A successful model achieves **high intra-cluster similarity** (contents of a cluster are very similar to each other) and **high inter-cluster separation** (contents of different clusters are very dissimilar).
*   **Standardization Rule:** Features with widely differing ranges **must be standardized to the same scale** before performing any clustering analysis.

### **2. Essential Formulas**
**Distance Metrics**
*   **Euclidean Distance (2D):** $d_{A, B} = \sqrt{(x_A - x_B)^2 + (y_A - y_B)^2}$.
*   **Manhattan Distance (2D):** $d_{A, B} = |x_A - x_B| + |y_A - y_B|$.

**K-Means Algorithms**
*   **New Centroid Calculation:** The new centroid is the mean of all included data records: $C' = (\frac{\sum x}{n}, \frac{\sum y}{n})$.
*   **Computation Complexity:** $O(M \cdot n \cdot K)$, where $M$ is the number of iterations, $n$ is the number of data records, and $K$ is the number of centroids.

**Gaussian Mixture Models (GMM) / Expectation Maximization (EM)**
*   **Initial Weight Constraint:** The assumed probability weights ($q_k$) for clusters must equal 1: $\sum_{k=1}^K q_k = 1$.
*   **Membership Likelihood (Responsibility) of $x_i$ for cluster $k$ ($r_{ik}$):** $r_{ik} = \frac{q_k \cdot P(x_i|k)}{\sum_{k=1}^K q_k \cdot P(x_i|k)}$.
*   **Total Likelihood for Cluster $k$ ($m_k$):** $m_k = \sum_{i=1}^n r_{ik}$.
*   **Parameter Updates (M-Step):** 
    *   Updated weight: $q_k' = \frac{m_k}{\sum_{k=1}^K m_k}$.
    *   Updated mean: $\mu_k' = \frac{1}{m_k} \sum_{i=1}^n r_{ik} \cdot x_i$.
    *   Updated standard deviation/covariance: $\sigma_k'^2 = \frac{1}{m_k} \sum_{i=1}^n r_{ik} \cdot (x_i - \mu_k')^T \cdot (x_i - \mu_k')$.

**Density-Based Spatial Clustering (DBSCAN)**
*   **$\epsilon$-neighborhood:** The set of points contained in a shape of extension $2\epsilon$, centered at a data record $p$: $N(p) = \{q \in d | dist(p,q) \leq \epsilon\}$.

### **3. Algorithm Rules & Relationships**

**K-Means (Centroid-Based)**
*   **Hard Clustering:** Each data record is assigned to **one and only one** cluster based on the minimum distance to a centroid.
*   **Workflow:** Operates in three steps: **Initialization** (randomly guessing $K$ centroids), **Assignment** (connecting records to the nearest centroid), and **Update** (recalculating the centroid as the mean point), repeating until convergence.

**GMM (Probabilistic Generative Models)**
*   **Soft Clustering:** Every data record has a probability of belonging to each cluster, and is finally assigned to the cluster with the highest membership probability.
*   **Workflow:** Uses the **Expectation Maximization (EM) algorithm** to alternate between the E-step (computing membership probability) and M-step (updating parameters) to find the maximum likelihood estimates. 

**Hierarchical Clustering**
*   **Linkage Rules (Distance between clusters):**
    *   **Single linkage:** Distance between the *closest* points of two clusters.
    *   **Complete linkage:** Distance between the *farthest* points of two clusters.
    *   **Average linkage:** Arithmetic mean of the distances between objects of two clusters.
*   **Optimal Clusters Rule:** To find the optimal number of clusters on a dendrogram, find the largest vertical line that does not intersect any other clusters, draw a horizontal line through it, and **count the number of intersections**.

**DBSCAN (Density-Based)**
*   **Point Classifications:**
    *   **Core point:** Its $\epsilon$-neighborhood contains at least the minimum density value (`MinPts`).
    *   **Border point:** Contains fewer than `MinPts` but falls within the $\epsilon$-neighborhood of a core point.
    *   **Noise point (Outlier):** Fits neither core nor border point classifications.

### **4. Points to Remember (Advantages & Limitations)**
*   **K-Means:** Requires you to guess the number of clusters ($K$) in advance, assumes all clusters are spherical, only handles numerical data, and fake outliers can skew cluster structures. It is fast for large datasets.
*   **GMM:** Best for complex, non-circular shapes (ellipsoids) and soft boundaries. However, the heavy computations of conditional probabilities can make the EM algorithm slow, and it can get stuck in a local maximum.
*   **Hierarchical:** Highly visual (via dendrograms) and does not require assuming the number of clusters at the start. However, **once a merge is made, it cannot be undone**, and it is highly sensitive to outliers and unsuited for large datasets.
*   **DBSCAN:** Automatically infers the number of clusters, identifies arbitrarily shaped clusters, and cleanly isolates noise/outliers. However, it **fails when clusters have varying densities** and can be very difficult to tune without domain knowledge.

# 3. ML Fundamentals of Regression Analysis and Modeling Methods

Here is a comprehensive cheat sheet covering the formulas, rules, and key points to remember about regression analysis, based on the sources provided:

### **1. Core Concepts & Rules**
*   **Definition:** Regression is a supervised learning method that determines the relationship between dependent (target) variables and independent variables.
*   **Simple vs. Multiple Regression:** A regression with only one independent variable is called a simple regression, while one with multiple independent variables is a multiple regression.
*   **Four Steps of Regression Analysis:** 
    1.  **Model Selection:** Choosing the relationship shape (linear or nonlinear, simple or multiple).
    2.  **Model Fitting:** Finding the unknown coefficients of the chosen model using the dataset.
    3.  **Model Prediction:** Using the model to estimate target variables for unseen data.
    4.  **Model Evaluation:** Checking how close predictions are to desired target values.

### **2. Linear & Nonlinear Regression**
*   **Linear Regression Formula (Multiple):** $y = a_1 \cdot x_1 + a_2 \cdot x_2 + \dots + a_n \cdot x_n + b$, where $a$ values are coefficients and $b$ is the bias or constant term.
*   **Nonlinear Regression Formula:** Modeled by a nonlinear combination, such as $y = a_1 \cdot x_1^2 + a_2 \cdot e^{x_2} + \dots + a_n \cdot x_n^3 + b$.
*   **Residue ($\epsilon$):** The difference between the predicted value ($y$) and the actual observation ($\hat{y}$). The formula is $y = \hat{y} + \epsilon$.
*   **Sum of Squared Errors (E):** The primary evaluation metric to minimize in simple linear regression. 
    *   **Formula:** $E = \sum (y - \hat{y})^2 = \sum (y - (a \cdot x + b))^2$.
*   **Calculating Optimum Coefficients ($a$ & $b$):** Derived by setting the partial derivatives of $E$ to zero ($\frac{\partial E}{\partial a} = 0$ and $\frac{\partial E}{\partial b} = 0$). 
    *   $b = \frac{1}{n} \sum y - a\frac{1}{n} \sum x$.
    *   $a = \frac{n\sum yx - \sum y\sum x}{n\sum x^2 - (\sum x)^2}$.

### **3. Evaluation Metrics & Linearity Rules**
*   **Coefficient of Determination ($R^2$):** Used to gauge the quality of fit by describing the proportion of variance in the dependent variable explained by the independent variables.
*   **$R^2$ Rule:** It ranges from zero to one, and a value closer to one indicates a better quality of fit.
*   **Test of Linearity:** For a model to satisfy the linearity assumption, the residue ($\epsilon$) values must be small compared to a threshold and must not display any significant pattern. If it fails, a nonlinear model should be used.

### **4. Logistic Regression**
*   **Use Case:** Used when the dependent variable follows a binomial distribution and takes on categorical or discrete binary values (e.g., "0" or "1", yes/no).
*   **How it Works:** It introduces an "S" shaped logit function to the linear regression equation to map predictions into exact binary values.
*   **Odds Ratio:** The probability of an event occurring divided by the probability of it not occurring.
    *   **Formula:** $\frac{p(y=1)}{1 - p(y=1)}$.
*   **Logit Function Formula:** Predicts the natural logarithm of the odds ratio: $\log(\frac{p}{1-p}) = a \cdot x + b$.
*   **Logistic Regression Equation:** $p(y=1) = \frac{1}{1 + e^{-(a \cdot x + b)}}$.

### **5. Quantile Regression**
*   **Use Case:** Ideal for skewed data, multimodal data, or data with outliers where looking at the "average" changes fails to capture the underlying data structure. 
*   **How it Works:** It divides the dependent variable into specific segments (quantiles like 5%, 10%, 50%, 90%) and calculates different regression coefficients for each level.
*   **Objective Function:** Instead of minimizing squared residuals, it minimizes a weighted sum of absolute errors for the specific quantile level ($\tau$).
    *   **Formula:** $\min \left( \tau \sum_{\text{above }\tau} |y - \hat{y}| + (1 - \tau) \sum_{\text{below }\tau} |y - \hat{y}| \right)$.

### **6. Regularization (Ridge & Lasso Regression)**
*   **Use Case:** Adds a penalty term to the regression model to prevent overfitting the training dataset and to mitigate the negative impact of outliers.
*   **Ridge Regression (L2 Regularization):** Penalizes the sum of the squares of the model coefficients ($W$), which mainly punishes the largest coefficients.
    *   **Formula:** $E = \sum (y - \hat{y})^2 + \lambda \sum W^2$.
    *   **Rule for $\lambda$:** Higher $\lambda$ (penalty level) places greater emphasis on reducing coefficient magnitudes, but tolerates higher residuals.
*   **Lasso Regression (L1 Regularization):** Minimizes the sum of the absolute values of the coefficients instead of their squares.
    *   **Formula:** $E = \sum (y - \hat{y})^2 + \lambda \sum |W|$.
    *   **Rule for sparsity:** Lasso drives down both large and small coefficients and preferentially sets some to exactly zero, heavily favoring sparse models.
*   **Tuning $\lambda$:** Selecting the optimum $\lambda$ requires a trade-off; you must investigate model performance over a range of values to find the one that best avoids overfitting.

# 5. ML Foundations and Methodologies of Decision Tree Classifiers

### **Mathematical Formulas**

**Classification Splitting Criteria:**
*   **Entropy (Impurity Measure):** $Info(D) = \sum_{k=1}^m -P_k \cdot \log P_k$, where $P_k$ is the probability that an instance belongs to class $k$.
*   **Entropy of a Partitioned Dataset:** $Info_f(D) = \sum_{i=1}^v \frac{|d_i|}{|D|} \cdot Info(d_i)$, which calculates the weighted sum of entropy for subsets partitioned by feature $f$.
*   **Information Gain:** $Gain(f) = Info(D) - Info_f(D)$, which represents the reduction in entropy achieved by splitting the dataset using feature $f$. 
*   **Split Info (Normalization Factor):** $SplitInfo_f(D) = -\sum_{i=1}^v \frac{|d_i|}{|D|} \cdot \log \frac{|d_i|}{|D|}$.
*   **Gain Ratio:** $gainRatio(f) = \frac{Gain(f)}{SplitInfo(f)}$, which is used to prevent the tree from overly favoring attributes with highly distinct values (like ID fields).
*   **Gini Impurity Index:** $Gini(D) = 1 - \sum_{k=1}^m P_k^2$.
*   **Gini Index for a Binary Split:** $Gini_f(D) = \frac{|d_1|}{|D|} Gini(d_1) + \frac{|d_2|}{|D|} Gini(d_2)$, which calculates the weighted sum of impurity for exactly two sets.
*   **Gini Impurity Reduction:** $\Delta Gini(f) = Gini(D) - Gini_f(D)$.

**Regression Splitting Criteria:**
*   **Standard Deviation Reduction:** $SDR(c,x) = S_c - S(c,x)$, where the variance/deviation of numerical targets is reduced by splitting instances via feature vector $x$.
*   **Coefficient of Variation:** $CV = \frac{S}{\bar{x}} \cdot 100\%$, which acts as a stopping condition for branching during regression tree induction.

---

### **Decision Tree Algorithms & Splitting Rules**

*   **ID3 (Iterative Dichotomiser 3):** Uses **Information Gain** to select the best features, which heavily penalizes heterogeneous data but only works natively with discrete, non-continuous features. For regression trees predicting numerical values, ID3 uses **Standard Deviation Reduction** instead.
*   **C4.5:** An extension of ID3 that uses **Gain Ratio** to rank features, overcoming ID3's flaw of over-ranking distinct identifier fields. Unlike ID3, C4.5 can deal with both continuous and discrete features, handles missing values, and supports post-pruning.
*   **CART (Classification and Regression Trees):** Uses the **Gini Index** to evaluate binary splits (splitting a node into exactly two child nodes) to maximize impurity reduction. 

---

### **Important Relationships & Points to Remember**

*   **Decision Rules:** Any path from the root node of the tree down to a leaf node represents a single, interpretable decision/classification rule (e.g., *If $f_1 < 0$ and $f_2 < 1$, then class = $C_1$*).
*   **Overfitting vs. Underfitting:** **Overfitting** occurs when trees become too complex by learning the training data too perfectly, thus losing the ability to reliably generalize to new data instances. **Underfitting** occurs when the tree is too simple or strictly pruned to classify instances accurately.
*   **Pruning:** To prevent overfitting and remove noisy data, trees are simplified by replacing unnecessary branches with a leaf node containing the most frequent class.
    *   **Pre-pruning:** Stops the tree-building process early by placing a leaf node if the ranking measure (like Information Gain or Gini index) falls below a predefined threshold.
    *   **Post-pruning:** Builds the full tree first, then evaluates the tree in a bottom-up manner, replacing decision nodes with leaf nodes if it improves the overall cost complexity (a function of error rate and leaf count).
*   **Weak Learners:** Decision trees are often restricted to a maximum depth or constrained to a single-level split (called a **decision stump**) to be purposefully used as "weak learners" in ensemble models.

---

### **Ensemble Methods Using Decision Trees**

*   **Bagging (Bootstrap Aggregation):** Forms multiple independent weak learners in parallel by taking random subsets (bootstrap samples) of the data. The final classification is decided by a simple unweighted majority vote from all learners.
*   **Random Forest:** A specialized bagging method that randomly restricts the subset of features available at each split, which efficiently handles high-dimensionality data and reduces the correlation between the individual trees.
*   **Boosting:** A sequential method where each new learner iteratively updates its training focus by giving a higher mathematical weight to the instances that were misclassified by the previous learner.

# 6. ML Genetic Algorithms: Principles and Evolutionary Optimization Phases

### **Genetic Algorithms Cheat Sheet**

#### **1. Key Definitions & Relationships**
*   **Population**: A group of objects in a particular space that represents a set of feasible solutions that may contain the optimal solution for a specific problem.
*   **Gene**: A heredity unit holding a characteristic. In computational terms, it is technically encoded as a bit representing a specific feature of a solution.
*   **Chromosome / DNA**: A sequence of genes. It is encoded as a list of bits, representing a complete possible solution (genotype) to a problem.
*   **Fitness Function**: An objective function used to measure how well a specific chromosome (solution) performs its designed aims in an environment. 
*   **Locus**: A specific point or location within a chromosome.

#### **2. The Five Phases of Evolutionary Optimization**
The genetic algorithm is an iterative searching process aiming to find successively better solutions through five distinct phases:
1.  **Initialization**: A random population of $n$ chromosomes is generated to serve as the initial set of solutions. This population should have a high coverage of the solution domain.
2.  **Evaluation (Fitness)**: The fitness of each individual chromosome is calculated using the fitness function, outputting a fitness score based on the problem's objective.
3.  **Selection**: Based on the principle of "survival of the fittest", individuals with fitness scores higher than a specific threshold are selected to breed the next generation. 
4.  **Crossover**: New offspring are created by mixing genes from the paired chromosomes of selected parents.
    *   **Single-point crossover**: The algorithm randomly chooses a locus and exchanges the subsequences before and after that point between two chromosomes.
    *   **Multi-point crossover**: A set of random points is selected, the chromosomes are cut at these points, and corresponding sections are swapped.
5.  **Mutation**: Random genes are flipped (from 1 to 0, or 0 to 1) at arbitrary locations in the offspring's chromosome based on a set probability. This prevents the population from falling into local optimum solutions too quickly.

#### **3. Important Rules & Points to Remember**
*   **Heuristic Nature**: Because genetic algorithms are a heuristic technique, they **cannot guarantee global optimality**. However, they provide near-optimum solutions that are usually sufficient for practical purposes.
*   **Stopping Conditions**: The evolutionary iteration stops when one of three conditions is met: 
    1. An acceptable level of fitness is achieved.
    2. No further improvement in fitness can be achieved.
    3. A pre-defined maximum number of iterations has been reached.
*   **Machine Learning Applications**: Genetic algorithms are widely used for feature selection (finding the subset of features that provides the best model performance) and hyperparameter tuning (e.g., limits on decision tree depth or number of clusters in k-means).
*   **Infeasible Solutions**: During evolution, solutions that break constraints are allowed to exist, but a mathematical penalty is imposed on their fitness score to reflect this.

#### **4. Formulas: The Knapsack Problem Application**
The Knapsack Problem is a classic optimization problem where you must maximize the total value of items placed in a knapsack without exceeding its maximum weight capacity, $M$.

**Variables:**
*   $x_j$: A binary gene where $x_j = 1$ if the item is selected for the knapsack, and $x_j = 0$ if it is left out.
*   $p_j$: The profit or value of item $j$.
*   $w_j$: The weight of item $j$.

**Formulas:**
-   **Base Fitness Function (Total Profit)**:
    $f(S) = \sum_{j=1}^{n} p_jx_j$
-   **Weight Constraint**: 
    $\sum_{j=1}^{n} w_jx_j \le M$
-   **Penalty Function** (For exceeding weight capacity):
    $Penalty = Q \cdot t(M - \sum_{j=1}^{n} w_jx_j)$

    -   *Where* $Q$ is a sufficiently large positive number (e.g., $Q = 6$).
    -   *Where* $t(z) = 0$ if $z \ge 0$.
    -   *Where* $t(z) = -z$ if $z < 0$.

-   **Adjusted Fitness Function** (Subtracting the penalty from the profit):
    $f(S) = \sum_{j=1}^{n} p_jx_j - Q \cdot t(M - \sum_{j=1}^{n} w_jx_j)$
